```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 
```

1. **Постройте график временного ряда , отображающий изменение температуры на протяжении недели. На графике должны быть отображены температуры на каждый день недели. Также небходимо дать корректные названия осям и самому графику (данное требование применимо ко всем заданиям)**

```python
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [20, 22, 25, 23, 24, 27, 26]

plt.figure(figsize=(10, 6))
plt.plot(weekdays, temperatures, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

plt.title('Изменение температуры на протяжении недели', fontsize=14)
plt.xlabel('День недели', fontsize=12)
plt.ylabel('Температура', fontsize=12)

plt.grid(True)
plt.show()
```

2. **Создайте `DataFrame` из предоставленных списков. Постройте scatterplot, отображающий зависимость размера скидки и количества проданных товаров. Необходимо выделить категории товаров разными маркерами и цветами. С данной задачей поможет параметр `hue`.**


```python
products = ['Shoes', 'Clothes', 'Accessories', 'Shoes', 'Clothes', 
            'Accessories', 'Shoes', 'Clothes', 'Accessories', 'Shoes', 'Clothes', 'Accessories']
discounts = [10, 25, 15, 20, 30, 18, 5, 12, 20, 30, 22, 27]
sales = [50, 75, 60, 55, 80, 70, 40, 65, 50, 75, 90, 85]


df = pd.DataFrame({
    'Product': products,
    'Discount': discounts,
    'Sales': sales
})


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Sales', hue='Product', style='Product', s=100)


plt.title('Зависимость размера скидки и количества проданных товаров', fontsize=14)
plt.xlabel('Размер скидки (%)', fontsize=12)
plt.ylabel('Количество проданных товаров', fontsize=12)

plt.grid(True)
plt.show()
```

3. **Постройте тепловую карту, отображающую матрицу корреляций для предложенного датасета. Воспользуйтесь методом `corr()` к `DateFrame`**

```python
df = pd.DataFrame({'col1': [1, 2, 7, 4, 56], 
                   'col2': [1, 19, 6, 13, -10], 
                   'col3': [2, 12, -9, 12, 15], 
                   'col4': [1, 8, 15, 16, 21],
                   'col5': [7, 4, 3, 17, -3]})

correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Тепловая карта корреляции для матрицы данных', fontsize=14)

plt.show()
```

4. **Постройте столбчатую диаграмму, отражающую количество продаж товара за последние 6 месяцев.**

```python
sales = pd.DataFrame({'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
                      'Sales': [350, 450, 600, 480, 750, 900]})

plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Sales', data=sales, palette='viridis', hue='Month')

plt.title('Количество продаж товара за последние 6 месяцев', fontsize=14)
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество продаж', fontsize=12)

plt.show()

```

5. **Необходимо построить boxplot, отображающий распределение значений для трех признаков в предложенном датасете.**

```python
data = pd.DataFrame({'col1': [1, 2, 3, 4, 5], 
                     'col2': [2, 4, 6, 8, 10], 
                     'col3': [3, 6, 9, 12, 15]})

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, palette='Set2')

plt.title('Распределение значений для трех признаков', fontsize=14)
plt.xlabel('Признаки', fontsize=12)
plt.ylabel('Значения', fontsize=12)

plt.show()
```

6. **Вам представлены 100 показаний температуры за последний месяц. Необходимо построить гистограмму, отображающую распределение данных.**

```python

np.random.seed(42)
data = pd.DataFrame({'Temperature': np.random.normal(15, 5, 100)})

plt.figure(figsize=(10, 6))
plt.hist(data['Temperature'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Распределение температуры за последний месяц', fontsize=14)
plt.xlabel('Температура', fontsize=12)
plt.ylabel('Частота', fontsize=12)

plt.show()
```

7. **Вам предоставлены данные о продажах продуктов в магазине по категориям товаров. Ваша задача - построить круговую диаграмму, отображающую, какую долю продаж в магазине составляет каждая категория товаров.**

```python

df = pd.DataFrame({
    'Category': ['Electronics', 'Apparel', 'Beauty', 'Grocery'],
    'Sales': [25000, 20000, 15000, 30000]
})

plt.figure(figsize=(8, 8))
plt.pie(df['Sales'], labels=df['Category'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], wedgeprops={'edgecolor': 'black'})

plt.title('Доля продаж по категориям товаров', fontsize=14)

plt.show()
```
